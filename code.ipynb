```python
from pynq import Overlay
import numpy as np
import cv2
import face_recognition
from pynq.lib.video import *
from pynq.overlays.base import BaseOverlay
base = BaseOverlay("base.bit")
```

```python
Mode = VideoMode(640,480,24)
hdmi_out = base.video.hdmi_out
hdmi_out.configure(Mode,PIXEL_BGR)
hdmi_out.start()
```

```python
frame_out_w = 640
frame_out_h = 480

frame_in_w = 640
frame_in_h = 480
```

```python
videoIn = cv2.VideoCapture(0)
videoIn.set(cv2.CAP_PROP_FRAME_WIDTH, frame_in_w);
videoIn.set(cv2.CAP_PROP_FRAME_HEIGHT, frame_in_h);

print("Capture device is open: " + str(videoIn.isOpened()))
```

```python
print("开始导入照片和名字")

shiyunjiao_image = face_recognition.load_image_file("ShiYunjiao.jpg")
shiyunjiao_face_encoding = face_recognition.face_encodings(shiyunjiao_image)[0]

direnjie_image = face_recognition.load_image_file("DiRenjie.jpg")
direnjie_face_encoding = face_recognition.face_encodings(direnjie_image)[0]

TomHolland_image = face_recognition.load_image_file("TomHolland.jpg")
TomHolland_face_encoding = face_recognition.face_encodings(TomHolland_image)[0]

BenedictCumberbatch_image = face_recognition.load_image_file("BenedictCumberbatch.jpg")
BenedictCumberbatch_face_encoding = face_recognition.face_encodings(BenedictCumberbatch_image)[0]

known_face_encodings = [
    shiyunjiao_face_encoding,
    direnjie_face_encoding,
    TomHolland_face_encoding,
    BenedictCumberbatch_face_encoding
]

known_face_names = [
    "ShiYunjiao",
    "DiRenjie",
    "TomHolland",
    "BenedictCumberbatch"
]

face_locations = []
face_encodings = []
face_names = []

print("导入成功")
```

```python
print("开始识别")

from pynq.lib.video import *
import asyncio 

process_this_frame = True

ret, frame = videoIn.read()

while True:
    ret, frame = videoIn.read()
    frame_1080p = np.zeros((600,1024,3)).astype(np.uint8)       
    inputBGR = cv2.resize(frame, (0,0),fx=0.25, fy=0.25)
    rgb_small_frame = inputBGR[:,:,::-1]
    
    if process_this_frame:
        face_locations = face_recognition.face_locations(rgb_small_frame)
        print("在图片里找到了{}张脸".format(len(face_locations)))
        face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)

        face_names = []
        for face_encoding in face_encodings:
            matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
            name = "UnknowPeople"

            if True in matches:
                first_match_index = matches.index(True)
                name = known_face_names[first_match_index]

            face_names.append(name)

    process_this_frame = not process_this_frame

    for (top, right, bottom, left), name in zip(face_locations, face_names):
        top *= 4
        right *= 4
        bottom *= 4
        left *= 4

        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left, bottom - 10), font, 1.0, (255, 255, 255), 1)

    outframe = hdmi_out.newframe()
    outframe[0:480,0:640,:] = frame[0:480,0:640,:]
    hdmi_out.writeframe(outframe)
```

```python
videoIn.release()
hdmi_out.stop()
del hdmi_out
```
